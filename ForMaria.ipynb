## Read in census tract shape files by wwtp
They are in a folder Census Tracts and WWTPs.  In those subfolders are a whole bunch of files with a whole bunch of extensions.  I read that even though it looks like you are only reading in the .shp file, it is actually using the whole set.

```python
wwfolders= os.listdir(os.path.join(filepath,'Census Tracts and WWTPs',''))
wwfiles = []
catchnames = []
for folder in wwfolders:
    allfiles = os.listdir(os.path.join(filepath,'Census Tracts and WWTPs',folder,''))
    wwfiles.append(os.path.join(filepath,'Census Tracts and WWTPs',folder,[
        myfile for myfile in allfiles if ('.shp' in myfile) and ('Census_Tract' in myfile)][0]))
    catchnames.append(folder.split(' ')[0])
cdf = pd.DataFrame()

for catch,myfile in zip(catchnames,wwfiles):
    shp = gpd.read_file(myfile)
    mydf = shp.copy()
    mydf['catch'] = catch
    if cdf.empty:
        cdf = mydf
    else:
        cdf = pd.concat([cdf,mydf], ignore_index=True)

cdf.TRACTCE = pd.to_numeric(cdf.TRACTCE)
cdf.rename(columns={'TRACTCE':'census_tract'},inplace=True)
cdf.INTPTLAT = pd.to_numeric(cdf.INTPTLAT)
cdf.INTPTLON = pd.to_numeric(cdf.INTPTLON)
# the next line is because for some reason there are some tracts in this file that are too far south...
cdf = cdf[cdf.INTPTLAT>41]
cdf.info()
```

```python
fig, ax = plt.subplots(figsize=(10,10))
divider = make_axes_locatable(ax)
# ax.set_xlim(-88.4,-87.4)
# ax.set_ylim(41.4,42.3)
cdf.plot(ax=ax,column='catch',legend=True)

plt.show()
```

## Read in Relationship Files
The zcta_tract_rel_10.csv file actually has a whole lot of information, we're only reading in some of it.  You may want to take a look, as some other might be useful in the future.

```python
zipcensus = pd.read_csv(os.path.join(filepath,'listfiles','zcta_tract_rel_10.csv'))
zipcensus = zipcensus[(zipcensus.STATE == 17) & (zipcensus.COUNTY == 31)][['ZCTA5','TRACT','TRPOP','ZPOP']].reset_index(drop=True).rename(
    columns={'ZCTA5':'zipcode','TRACT':'census_tract','TRPOP':'ctpop','ZPOP':'zippop'})
zipcensus.info()

ctpop = zipcensus[['census_tract','ctpop']].sort_values('census_tract').drop_duplicates()
ctpop.info()

zpop = zipcensus[['zipcode','zippop']].sort_values('zippop').drop_duplicates()
zpop.info()

ccatch = pd.read_csv(os.path.join(filepath,'listfiles','census_catchment.csv'))
ccatch.info()

ZipCatch = pd.read_csv(os.path.join(filepath,'listfiles','ZipWWTP.csv'))
catches = ZipCatch.columns.tolist()
zcatch = pd.DataFrame()
for n,catch in enumerate(catches):
    ZipCatch = pd.read_csv(os.path.join(filepath,'listfiles','ZipWWTP.csv'),usecols=[catch]).dropna().astype('int64')
    mydf = pd.DataFrame({'zipcatch':catch,'zipcode':ZipCatch[catch].tolist()})
    if zcatch.empty:
        zcatch = mydf
    else:
        zcatch = pd.concat([zcatch,mydf], ignore_index=True)
zcatch.info()
```

```python
fig, ax = plt.subplots(figsize=(15,15))
cdf.plot(ax=ax,column='catch',legend=True,alpha=.5,edgecolor='w')
zshp.boundary.plot(ax=ax,edgecolor='k')
cshp.plot(ax=ax,edgecolor='dimgrey',color= "darkgrey",alpha=.7)

tcatch = 'Stickney'
usezips = zcatch[zcatch.zipcatch == tcatch].zipcode.unique().tolist()
usesomezips = [myzip for myzip in usezips if myzip not in zcatch[~(
    zcatch.zipcatch == tcatch)].zipcode.unique()]
usects = zipcensus[(zipcensus.zipcode.isin(usezips)) & (zipcensus.census_tract.isin(cshp.census_tract))].census_tract.unique().tolist()
usesomects = [myct for myct in zipcensus[(zipcensus.zipcode.isin(
    usesomezips)) & (zipcensus.census_tract.isin(cshp.census_tract))].census_tract.unique() if myct not in zipcensus[~(zipcensus.zipcode.isin(
    usesomezips))].census_tract.unique()]
usecatchcts = ccatch[(ccatch.catchment==tcatch) & (ccatch.census_tract.isin(cshp.census_tract))].census_tract.unique().tolist()

mypart = cdf[(cdf.census_tract.isin(usects))].dissolve(by='COUNTYFP')
mypart.boundary.plot(ax=ax,edgecolor='green',linewidth=4)
cdf[(cdf.census_tract.isin(usects))].plot(ax=ax, color='tab:green',edgecolor='darkgrey',alpha=.4)
mypart = cdf[(cdf.census_tract.isin(usecatchcts))].dissolve(by='COUNTYFP')
mypart.boundary.plot(ax=ax,edgecolor='orange',linewidth=4)
cdf[(cdf.census_tract.isin(usecatchcts))].plot(ax=ax, color='tab:orange',alpha=.4)
mypart = cdf[(cdf.census_tract.isin(usesomects))].dissolve(by='COUNTYFP')
mypart.boundary.plot(ax=ax,edgecolor='blue',linewidth=4)
cdf[(cdf.census_tract.isin(usesomects))].plot(ax=ax, color='tab:blue',alpha=.4)

plt.gca().set_title('Focus on Stickney (three "subsets")',y=.97,fontsize=20)
plt.show()
```

## Submaps and neighbors
First, add a column that lists the zips that share a side or vertice.

```python
# add NEIGHBORS column
zshp["neighbors"] = None  
for index, country in zshp.iterrows():   
    # get 'not disjoint' countries
    neighbors = zshp[zshp.geometry.touches(country.geometry)].zipcode.unique().tolist()

    # remove own name of the country from the list
    neighbors = [ name for name in neighbors if country.zipcode != name ]
    # The following if statement is because for some reason some of them don't compute correctly.
    if country['zipcode'] == 60614:
        neighbors.append(60622)
        neighbors.append(60642)
    elif country['zipcode'] == 60622:
        neighbors.append(60614)
    elif country['zipcode'] == 60642:
        neighbors.append(60614)
        neighbors.append(60654)
    elif country['zipcode'] == 60654:
        neighbors.append(60642)
        neighbors.append(60611)
    # add names of neighbors as NEIGHBORS value
    zshp.at[index, "neighbors"] = neighbors
```
